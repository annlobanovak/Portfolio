2. **Анализ деятельности компании**

-Кейс:  Pens and Pencils -  компания специализирующаяся на B2B и B2C-продажах офисных товаров.

-Задача: Проанализировать работу компании с точки зрения ее эффективности и дать рекомендации по масштабированию бизнеса, а именно в каком штате лучше открыть офлайн-магазин.

2. **Анализ бизнес модели онлайн магазина**

-Кейс:  Новосибирская компания специализирующаяся на онлайн продаже экотоваров планирует расширение бизнеса в другие регионы

-Задача: Проанализировать дейятельность компании, эффективность бизнес модели, дать прогнозы и рекомендации по масштабированию бизнеса.

-Результат: Я провела маркетинговый и продуктовый анализ бизнеса. Используя когортный анализ и показатели юнит-экономики, сформулировала основные выводы о рентабельности, определила основные источники привлечения и самые успешные регионы. Дала рекомендации по расширению в другие регионы. Работа реализована и визуализирована в GoogleSheets.
